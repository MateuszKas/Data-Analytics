```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
#Part 1 and Part 2
df=pd.read_csv(r'./Data1.csv', index_col=0)
df.index.name='Dates'
```

```python
#Part 3
ax = df.plot(figsize = (15,6))
df.plot(subplots = True, figsize = (15,6), legend = None)
plt.plot(df)
```

```python
#Part 4
pd_melt=pd.melt(df, id_vars =['theta_1'], value_vars = ['theta_2','theta_3','theta_4','theta_5','theta_6'])
g = sns.FacetGrid(pd_melt, sharex = False)
g.map(plt.hist, "value");
```
