```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sn
import numpy as np
import seaborn as sns
import scipy as sp
```

```python
"""""
Exercise 1 - Normal Distribution
Data Frame contains normal distribution with different values of mean and variance
"""""
normal_distribution = {
    "default":np.random.normal(1,1,10000),
    "mean2_var1":np.random.normal(2,1,10000),
    "mean1_var10":np.random.normal(1,10,10000),
    "mean5_var5":np.random.normal(5,5,10000),
}
df = pd.DataFrame(normal_distribution)
#print(df)
df_melt = pd.melt(df, var_name='Parameters',value_name='Value')
#print(df_melt)
g = sns.FacetGrid(df_melt,col="Parameters",sharex=True)
g.map(plt.hist,'Value',cumulative=False,bins=512)

g = sns.FacetGrid(df_melt,col="Parameters",sharex=True)
g.map(plt.hist,'Value',cumulative=False,bins=64)

g = sns.FacetGrid(df_melt,col="Parameters",sharex=True)
g.map(plt.hist,'Value',cumulative=False,bins=8)

g = sns.FacetGrid(df_melt,col="Parameters",sharex=True)
g.map(plt.hist,'Value',cumulative=True,bins=64)
"""""
Graphs displayed below shows histograms for normal distribution generated above. As we can see below, 
if we will not choose variancy and mean value on proper value histogram is hard to analyze. It is also 
hard to analyze histograms with small amount of samples, and that is the reason why we are generating 10000 
samples in each case. Value of bins parameter in map function determine estimation of graph (In first line 
we can observe that graph is not that smooth as in second line, but if we decrease bins number too much as 
in third line we will lose accuracy). We can also change parameter "cumulative" in map function for "True" 
(Displayed in fourth line).
"""""
```

```python
"""""
Exercise 2 - Inverse Logit
We transform earlier generated normal distribution with inverse logit using function fron scipy library

"""""
expit_df=sp.special.expit(df)
expit_dfmelt = pd.melt(expit_df,var_name='Parameters',value_name='Value')
#print(expit_dfmelt)
g = sns.FacetGrid(expit_dfmelt,col="Parameters",sharex=True)
g.map(plt.hist,'Value',cumulative=False,bins=64)

g = sns.FacetGrid(expit_dfmelt,col="Parameters",sharex=True)
g.map(plt.hist,'Value',cumulative=True,bins=64)
```

```python
"""""
Exercise 3 - Poisson distribution
DataFrame contains poisson distribution 

"""""
poisson_distribution = {
    "default":np.random.poisson(size=10000),
    "expectation of interval 5":np.random.poisson(lam=5,size=10000),
    "expectation of interval 10":np.random.poisson(lam=10,size=10000),
}
df_poisson = pd.DataFrame(poisson_distribution)
#print(df_poisson)
df_poisson_melt = pd.melt(df_poisson, var_name='parameters',value_name='Value')
#print(df_poisson_melt)
g = sns.FacetGrid(df_poisson_melt,col="parameters",sharex=True)
g.map(plt.hist,'Value',cumulative=False,bins=64)

g = sns.FacetGrid(df_poisson_melt,col="parameters",sharex=True)
g.map(plt.hist,'Value',cumulative=True,bins=64)
"""""
In this case standard deviation is equal to the square root of the expected value. So with increase of expected value, graph getting flatter.
"""""
```

```python
"""""
Exercise 4 - Beta distribution
DataFrame contains Beta distribution defined in interval (0,1)

"""""
beta_distribution = {
    "a=1,b=1":np.random.beta(a=1,b=1,size=10000),
    "a>1,b<1":np.random.beta(a=4,b=0.4,size=10000),
    "a<1,b>1":np.random.beta(a=0.6,b=4,size=10000),
    "a<1,b<1":np.random.beta(a=0.4,b=0.9,size=10000),
    "a>1,b>1":np.random.beta(a=2,b=4,size=10000),
    
}
df_beta = pd.DataFrame(beta_distribution)
#print(df_poisson)
df_beta_melt = pd.melt(df_beta, var_name='parameters',value_name='Value')
#print(df_poisson_melt)
g = sns.FacetGrid(df_beta_melt,col="parameters",sharex=True)
g.map(plt.hist,'Value',cumulative=False,bins=64)

g = sns.FacetGrid(df_beta_melt,col="parameters",sharex=True)
g.map(plt.hist,'Value',cumulative=True,bins=64)
"""""
We can control shape of graph by using different values of alpha and beta parameters. 
Expected value is equal alpha/(alpha+beta). On histograms below we can observe how shape 
change depends of different values of parameters.
"""""
```

```python
"""""
Exercise 5 - Log-normal distribution
DataFrame contains Log-normal distribution defined based on mean and standard deviation value

"""""
log_normal_distribution = {
    "default":np.random.lognormal(size=10000),
    "mean4_sigma1":np.random.lognormal(mean=4,sigma=1,size=10000),
    "mean1_sigma0.4":np.random.lognormal(mean=1,sigma=0.4,size=10000),
    "mean4_sigma0.1":np.random.lognormal(mean=4,sigma=0.1,size=10000),
    "mean1_sigma0.1":np.random.lognormal(mean=1,sigma=0.1,size=10000),


    
}
df_lognormal = pd.DataFrame(log_normal_distribution)
#print(df_poisson)
df_lognormal_melt = pd.melt(df_lognormal, var_name='parameters',value_name='Value')
#print(df_poisson_melt)
g = sns.FacetGrid(df_lognormal_melt,col="parameters",sharex=True)
g.map(plt.hist,'Value',cumulative=False,bins=64)

g = sns.FacetGrid(df_lognormal_melt,col="parameters",sharex=True)
g.map(plt.hist,'Value',cumulative=True,bins=64)
"""""
This is similar distribution to normal, but it does not contains negative values
"""""
```
